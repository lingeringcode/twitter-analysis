```python
import pandas as pd
from pandas.tseries.resample import TimeGrouper
from pandas.tseries.offsets import DateOffset
from pandas.tseries.offsets import *
rsa_kw = pd.read_csv('/home/lingeringcode/projects/rsa-analysis/material_rsaCount.csv')
rsa_kw['time'] = pd.to_datetime(pd.Series(rsa_kw['time']))
rsa_kw.set_index('time', drop=False, inplace=True)
rsa_kw.index = rsa_kw.index - DateOffset(hours = 5)
rsa_kw.index
```

```python
rsa_kw.head()
```

```python
rsa_kw.describe()
```

```python
rsa_kw5m = rsa_kw['time'].resample('5t', how='count')
rsa_kw5m.head()
```

```python
import vincent
vincent.core.initialize_notebook()
area = vincent.Area(rsa_kw5m)
area.legend(title='RSA14: Material')
area.axis_titles(x='Time', y='# of tweets with \'material\'/5 Minutes')
area.colors(brew='Spectral')
area.display()
```

```python
rsa_kw5m = rsa_kw['time'].resample('5Min', how='count')
rsa_kw5m.head()
```

```python
import vincent
vincent.core.initialize_notebook()
area = vincent.Area(rsa_kw5m)
area.legend(title='RSA14: Material')
area.axis_titles(x='Time', y='# of tweets with \'material\'/5 min')
area.colors(brew='Spectral')
area.display()
```
